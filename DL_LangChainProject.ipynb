```python
import os
from google.colab import userdata

# 1. Setup API Key (Enter it in the 🔑 Secrets tab on the left)
os.environ["OPENAI_API_KEY"] = userdata.get('openai_key')

# 2. Modern 2026 Imports
# Note: 'create_agent' now handles both the logic and the execution runtime
from langchain_openai import ChatOpenAI
from langchain_community.tools import DuckDuckGoSearchRun
from langchain.agents import create_agent

# 3. Initialize the Model (gpt-4o-mini is standard for agents)
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)

# 4. Define the Web Search Tool
search_tool = DuckDuckGoSearchRun()
tools = [search_tool]

# 5. Create the Agent
# In v1.2+, 'create_agent' returns a runnable object you can invoke immediately.
# It uses the ReAct pattern by default when tools are provided.
agent = create_agent(
    model=llm,
    tools=tools,
    system_prompt="You are a helpful research assistant. Use the search tool for current events."
)

# 6. Ask your question
print("--- Starting Agent ---")
try:
    # Use the standard 'messages' input format for modern agents
    response = agent.invoke({"messages": [("user", "What are the 3 most popular movies in theaters right now?")]})

    # Extract the last message from the agent (the answer)
    print("\n--- Final Answer ---")
    print(response["messages"][-1].content)
except Exception as e:
    print(f"Error: {e}")
```

```python
import os
from google.colab import userdata


os.environ["NVIDIA_API_KEY"] = userdata.get('nvidia_key')

from langchain_nvidia_ai_endpoints import ChatNVIDIA
from langchain_community.tools import DuckDuckGoSearchRun
from langchain.agents import create_agent


llm = ChatNVIDIA(
    model="meta/llama-3.1-70b-instruct",
    temperature=0
)

# 4. Web Search Tool
search_tool = DuckDuckGoSearchRun()
tools = [search_tool]

# 5. Create Agent
agent = create_agent(
    model=llm,
    tools=tools,
    system_prompt="You are a helpful research assistant. Use the search tool for current events."
)

# 6. Ask Question
print("--- Starting Agent ---")

try:
    response = agent.invoke({
        "messages": [
            ("user", "what is todays ipl match")
        ]
    })

    print("\n--- Final Answer ---")
    print(response["messages"][-1].content)

except Exception as e:
    print(f"Error: {e}")
```
